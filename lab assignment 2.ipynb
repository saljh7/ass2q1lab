```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Psin(x,n):
    terms = np.zeros(n+1)
    for k in range(n+1):
        numerator = ((-1) ** k) * (x ** ((2 * k) + 1))
        denominator = math.factorial((2 * k) + 1)
        terms[k] = numerator/denominator
    return terms.sum()
```

n = 10

The following function takes an input value x and return an array of the log errors for each n. for example at the zero entry is log(abs|psin(x,0)- sin(x)|)

```python
def logerrors(x):
    narray = np.arange(10)
    errorarray = np.empty(10)
    for entry in narray:
     computedvalue = Psin(x, entry)
     actualvalue = math.sin(x)
     if (computedvalue - actualvalue) != 0:
      errorarray[entry] = (math.log(abs(computedvalue - actualvalue)))
    return errorarray
```

i use x values from 10 degrees to 90 degrees in steps of 10, or pi/18 - pi/2

```python
xvalues = np.arange(math.pi/18, math.pi/2 + math.pi/18, math.pi/18)
narray = np.arange(10)
errorarray = np.empty(10)
for i in range(len(xvalues)):
     plt.plot(narray, logerrors(xvalues[i]))
plt.xlabel("N")
plt.ylabel("log|error|")
plt.show()
```

the slope represents the decrease in error for additional term in our power series. The sharp turining points represent where our series converges to sin(x).
